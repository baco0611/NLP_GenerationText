```python
import os
from pyvi import ViTokenizer
from sklearn.model_selection import train_test_split
import string
```

```python
def load_data(directory_path):
    all_texts = []
    count = 1
    for file_name in os.listdir(directory_path):
        print(count, "Loading in file", file_name)
        count+=1

        file_path = os.path.join(directory_path, file_name)
        with open(file_path, 'r', encoding='utf-16') as file:
            text = file.read()
            all_texts.append(text)
    return all_texts
```

```python
def clean_document(doc):
    doc = ViTokenizer.tokenize(doc) #Pyvi Vitokenizer library
    doc = doc.lower() #Lower
    tokens = doc.split() #Split in_to words
    table = str.maketrans('', '', string.punctuation.replace("_", "")) #Remove all punctuation
    tokens = [w.translate(table) for w in tokens]
    tokens = [word for word in tokens if word]
    return tokens
```

```python
def tokenize_texts(texts):
    return [ViTokenizer.tokenize(text) for text in texts]
```

```python
texts = load_data('./Train_Full/VanHoa')
```

```python
processed_texts = [clean_document(text) for text in texts]
```

```python
import pickle
with open('./0.VanHoaFull/processed_texts.pkl', 'wb') as f:
    pickle.dump(processed_texts, f)
```
